```python
# Importing libraries
import seaborn as sns
import numpy as np 
```

```python
# Reading the dataset 
iris = sns.load_dataset('iris')
iris
```

```python
# Scatterplot 
sns.scatterplot(x = 'sepal_length', y = 'petal_length', data = iris)
```

```python
# Lineplot 
sns.lineplot(x = 'sepal_length', y = 'petal_length', data = iris, ci = None)
```

```python
# Distplot 
sns.distplot(iris['petal_length'])
```

```python
# Barplot 
sns.barplot(x = 'species', y = 'petal_length', data = iris)
```

```python
# Countplot 
sns.countplot(x = 'petal_width', data = iris)
```

```python
# Box plot 
sns.boxplot(x = 'sepal_width', y = 'petal_width', data = iris)
```

```python
# Catplot 
sns.catplot(x = 'species', y = 'petal_length', hue = 'petal_width', data = iris)
```

```python
# Correlation
iris.corr()
```

```python
# Heatmap 
sns.heatmap(iris.corr(), annot = True)
```
