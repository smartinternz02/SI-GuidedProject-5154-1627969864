# Bank data ANN (Categorical)

```python
# 1. Importing the libraries 
import numpy as np 
import pandas as pd 
```

```python
# 2. Reading the datasets 
data = pd.read_csv('bank.csv')
```

```python
data.head(30)
```

```python
# 3. Handling missing values 
data.isnull().any() 
```

```python
# 4. Splitting data into inputs and outputs
inputs = data.iloc[:, :16].values
outputs = data.iloc[:, -1:].values
```

```python
inputs[0:2]
```

```python
outputs.shape
```

```python
outputs
```

```python
# Importing libraries to handle the categorical data 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
```

```python
# 5. Handling categorical data 
ct = ColumnTransformer([('one', OneHotEncoder(), [1,2,3,4,6,7,8,10,15])], remainder = 'passthrough')
inputs = ct.fit_transform(inputs)

cto = ColumnTransformer([('one', OneHotEncoder(), [0])], remainder = 'passthrough')
outputs = cto.fit_transform(outputs)
```

```python
inputs.shape
```

```python
inputs
```

```python
outputs
```

```python
# Importing libraries to handle splitting 
from sklearn.model_selection import train_test_split
```

```python
# 6. Splitting data to train and test sets
x_train,x_test,y_train,y_test = train_test_split(inputs,outputs,test_size = 0.2, random_state = 0)
```

```python
print('Inputs shape ', inputs.shape)
print('Outputs shape ', outputs.shape)
print('x_train shape ', x_train.shape)
print('x_test shape ', x_test.shape)
print('y_train shape ', y_train.shape)
print('y_test shape', y_test.shape)
```

```python
# Importing libraries to handle feature scaling 
from sklearn.preprocessing import StandardScaler
```

```python
# 7. Applying feature scaling 
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
x_train
```

```python
x_test
```

```python
# Import deep learning libraries
from tensorflow.keras.models import Sequential  
from tensorflow.keras.layers import Dense 
```

```python
categorical = Sequential() 
```

```python
# Input layer
categorical.add(Dense(units = 51, kernel_initializer = 'random_uniform', activation = 'relu'))
```

```python
# Hidden layer 1  
categorical.add(Dense(units = 102, kernel_initializer = 'random_uniform', activation = 'relu'))
```

```python
# Hidden layer 2
categorical.add(Dense(units = 102, kernel_initializer = 'random_uniform', activation = 'relu'))
```

```python
# Output layer 
categorical.add(Dense(units = 2, kernel_initializer = 'random_uniform', activation = 'softmax'))
```

```python
# Compiling model  
categorical.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['accuracy'])
```

```python
# Training the model 
categorical.fit(x_train, y_train, epochs = 150)
```

```python
# Testing the model with test set 
ypred = categorical.predict(x_test)
```
